```python
import language_tool_python
tool = language_tool_python.LanguageTool('en-US')
```

```python
import pandas as pd
```

```python
data=pd.read_excel("review_data.xlsx")
```

```python
grammer=data.text
```

```python
grammer=list(grammer)
```

```python
pip install googletrans==3.1.0a0
```

```python
from googletrans import Translator

translator = Translator()
for each in grammer:


  translated_text = translator.translate(each)
  word=translated_text.text
  print(tool.correct(word))
```
